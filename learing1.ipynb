```python
import numpy as np
a=np.arange(2 * 2 * 4).reshape(2, 2, 4)
print(a.nbytes)#占用的内存大小
print(a.size*a.itemsize)#内存大小
print(a.itemsize)#每个元素占用的字节数
print(a.size)#返回原数的个数
print(a.shape)#形状
print(a.dtype)#数据类型
```

```python
import numpy as np
rng=np.random.default_rng(seed=42)
nums1=rng.uniform(low=10,high=20,size=(5,5))
nums2=rng.integers(low=0,high=10,size=(5,5))
print(nums1)
print(nums2)
```

```python
import numpy as np
rng=np.random.default_rng(seed=42)
X=rng.integers(1,30,19)
X.mean()#求和
X.prod()#元素的积
X.std()#标准差
X.var()#方差
X.argmax()#最小值索引
X.argsort()#按位排序
np.median(X)#这个需要函数来算
```
