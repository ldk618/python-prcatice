```python
#归一化的一种方法
import numpy as np
rng=np.random.default_rng(seed=42)
Z=rng.integers(low=0,high=10,size=(5,5))
Z_norm = (Z - Z.min()) / (Z.max() - Z.min())#[min, max] → [0, max-min]// X`=X-Xmin/Xmax-Xmin
#这样就把范围缩小到[0,1]之间
print("原始范围:", Z.min(), "to", Z.max())
print("归一化后:", Z_norm.min(), "to", Z_norm.max())  # 0.0 to 1.0
```

```python
#Z-Score 标准化
import numpy as np
rng=np.random.default_rng(seed=42)
Z=rng.integers(low=0,high=10,size=(5,5))
Z_std=(Z-Z.mean())/Z.std()
print("均值",Z_std.mean())
print("标准化",Z_std.std())
```

```python
rng=np.random.default_rng(seed=42)
Z = rng.random((5, 5))
# 每行归一化（使每行向量长度为1）
Z_l2 = Z / np.linalg.norm(Z, axis=1, keepdims=True)#确保广播运算正常进行
row_norms = np.linalg.norm(Z_l2, axis=1)
print(Z_l2)
print(row_norms)
```
